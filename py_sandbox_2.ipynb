```python
earthquake_data = {
    'eq1':  {'Name': 'Northridge',    'Year': 1994, 'PGA': 0.45, 'AffectedCities': ['Los Angeles', 'San Fernando']},
    'eq2':  {'Name': 'Loma Prieta',   'Year': 1989, 'PGA': 0.38, 'AffectedCities': ['Santa Cruz', 'San Francisco']},
    'eq3':  {'Name': 'Kobe',          'Year': 1995, 'PGA': 0.42, 'AffectedCities': ['Kobe', 'Osaka', 'Kyoto']},
    'eq4':  {'Name': 'Chi-Chi',       'Year': 1999, 'PGA': 0.37, 'AffectedCities': ['Taipei', 'Taichung']},
    'eq5':  {'Name': 'Haiti',         'Year': 2010, 'PGA': 0.31, 'AffectedCities': ['Port-au-Prince']},
    'eq6':  {'Name': 'Sumatra',       'Year': 2004, 'PGA': 0.44, 'AffectedCities': ['Banda Aceh', 'Medan', 'Padang']},
    'eq7':  {'Name': 'Mexico City',   'Year': 1985, 'PGA': 0.36, 'AffectedCities': ['Mexico City']},
    'eq8':  {'Name': 'Iran East',     'Year': 1978, 'PGA': 0.29, 'AffectedCities': ['Tabas']},
    'eq9':  {'Name': 'Alaska',        'Year': 1964, 'PGA': 0.39, 'AffectedCities': ['Anchorage']},
    'eq10': {'Name': 'Nepal',         'Year': 2015, 'PGA': 0.33, 'AffectedCities': ['Kathmandu', 'Pokhara']},
}

def strong_earthquake_generator(data, threshold):
    for quake_id, details in data.items():
        if details['PGA'] >= threshold:
            yield quake_id, details

def print_strong_earthquake_report(data, threshold):
    count = 0
    for quake_id, details in strong_earthquake_generator(data, threshold):
        count += 1
        many_cities = len(details['AffectedCities']) > 2
        print(f"Earthquake {details['Name']} in {details['Year']} (ID: {quake_id}):")
        print(f"  - PGA: {details['PGA']:.2f}g")
        print(f"  - Number of affected cities: {len(details['AffectedCities'])}")
        print(f"  - Affected many cities? {'Yes' if many_cities else 'No'}\n")
    if count == 0:
        print("No strong earthquakes found.")

# Example usage:
threshold_value = 0.43
print_strong_earthquake_report(earthquake_data, threshold_value)
```

```python
# ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸÅÿ±Ÿàÿ¥⁄ØÿßŸá (ÿ≥ŸÅÿßÿ±ÿ¥‚ÄåŸáÿß)
sales_data = {
    'order1': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 2, 'Delivered': True},
    'order2': {'Product': 'Mouse', 'Price': 25, 'Quantity': 5, 'Delivered': False},
    'order3': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 3, 'Delivered': True},
    'order4': {'Product': 'Monitor', 'Price': 300, 'Quantity': 1, 'Delivered': True},
    'order5': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': False},
    'order6': {'Product': 'Mouse', 'Price': 25, 'Quantity': 2, 'Delivered': True},
    'order7': {'Product': 'Chair', 'Price': 85, 'Quantity': 1, 'Delivered': True},
    'order8': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': False},
    'order9': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': True},
    'order10': {'Product': 'Monitor', 'Price': 300, 'Quantity': 2, 'Delivered': False},
    'order11': {'Product': 'Mouse', 'Price': 25, 'Quantity': 3, 'Delivered': True},
    'order12': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 2, 'Delivered': False},
    'order13': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 4, 'Delivered': True},
    'order14': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': True},
    'order15': {'Product': 'Monitor', 'Price': 300, 'Quantity': 3, 'Delivered': False},
    'order16': {'Product': 'Chair', 'Price': 85, 'Quantity': 2, 'Delivered': True},
    'order17': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': False},
    'order18': {'Product': 'Mouse', 'Price': 25, 'Quantity': 4, 'Delivered': True},
    'order19': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 5, 'Delivered': False},
    'order20': {'Product': 'Monitor', 'Price': 300, 'Quantity': 2, 'Delivered': True},
    'order21': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': False},
    'order22': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 3, 'Delivered': True},
    'order23': {'Product': 'Chair', 'Price': 85, 'Quantity': 3, 'Delivered': True},
    'order24': {'Product': 'Mouse', 'Price': 25, 'Quantity': 1, 'Delivered': False},
    'order25': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 4, 'Delivered': True},
    'order26': {'Product': 'Monitor', 'Price': 300, 'Quantity': 1, 'Delivered': False},
    'order27': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 2, 'Delivered': True},
    'order28': {'Product': 'Desk', 'Price': 200, 'Quantity': 2, 'Delivered': True},
    'order29': {'Product': 'Chair', 'Price': 85, 'Quantity': 1, 'Delivered': False},
    'order30': {'Product': 'Mouse', 'Price': 25, 'Quantity': 5, 'Delivered': True},
    'order31': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 3, 'Delivered': False},
    'order32': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 5, 'Delivered': True},
    'order33': {'Product': 'Monitor', 'Price': 300, 'Quantity': 4, 'Delivered': False},
    'order34': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': True},
    'order35': {'Product': 'Chair', 'Price': 85, 'Quantity': 2, 'Delivered': False},
    'order36': {'Product': 'Mouse', 'Price': 25, 'Quantity': 2, 'Delivered': True},
    'order37': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 1, 'Delivered': True},
    'order38': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 3, 'Delivered': False},
    'order39': {'Product': 'Monitor', 'Price': 300, 'Quantity': 2, 'Delivered': True},
    'order40': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': False},
    'order41': {'Product': 'Chair', 'Price': 85, 'Quantity': 4, 'Delivered': True},
    'order42': {'Product': 'Mouse', 'Price': 25, 'Quantity': 3, 'Delivered': False},
    'order43': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': True},
    'order44': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 5, 'Delivered': False},
    'order45': {'Product': 'Monitor', 'Price': 300, 'Quantity': 3, 'Delivered': True},
    'order46': {'Product': 'Desk', 'Price': 200, 'Quantity': 2, 'Delivered': True},
    'order47': {'Product': 'Chair', 'Price': 85, 'Quantity': 1, 'Delivered': False},
    'order48': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 4, 'Delivered': True},
    'order49': {'Product': 'Mouse', 'Price': 25, 'Quantity': 2, 'Delivered': False},
    'order50': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 3, 'Delivered': True},
    'order51': {'Product': 'Monitor', 'Price': 300, 'Quantity': 1, 'Delivered': False},
    'order52': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': True},
    'order53': {'Product': 'Chair', 'Price': 85, 'Quantity': 5, 'Delivered': True},
    'order54': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 2, 'Delivered': False},
    'order55': {'Product': 'Mouse', 'Price': 25, 'Quantity': 4, 'Delivered': True},
    'order56': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 1, 'Delivered': False},
    'order57': {'Product': 'Monitor', 'Price': 300, 'Quantity': 2, 'Delivered': True},
    'order58': {'Product': 'Desk', 'Price': 200, 'Quantity': 3, 'Delivered': False},
    'order59': {'Product': 'Chair', 'Price': 85, 'Quantity': 2, 'Delivered': True},
    'order60': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 3, 'Delivered': False},
    'order61': {'Product': 'Mouse', 'Price': 25, 'Quantity': 1, 'Delivered': True},
    'order62': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 4, 'Delivered': True},
    'order63': {'Product': 'Monitor', 'Price': 300, 'Quantity': 5, 'Delivered': False},
    'order64': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': True},
    'order65': {'Product': 'Chair', 'Price': 85, 'Quantity': 3, 'Delivered': False},
    'order66': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': True},
    'order67': {'Product': 'Mouse', 'Price': 25, 'Quantity': 5, 'Delivered': False},
    'order68': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 2, 'Delivered': True},
    'order69': {'Product': 'Monitor', 'Price': 300, 'Quantity': 1, 'Delivered': True},
    'order70': {'Product': 'Desk', 'Price': 200, 'Quantity': 2, 'Delivered': False},
    'order71': {'Product': 'Chair', 'Price': 85, 'Quantity': 4, 'Delivered': True},
    'order72': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 2, 'Delivered': False},
    'order73': {'Product': 'Mouse', 'Price': 25, 'Quantity': 3, 'Delivered': True},
    'order74': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 1, 'Delivered': False},
    'order75': {'Product': 'Monitor', 'Price': 300, 'Quantity': 4, 'Delivered': True},
    'order76': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': True},
    'order77': {'Product': 'Chair', 'Price': 85, 'Quantity': 2, 'Delivered': False},
    'order78': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 5, 'Delivered': True},
    'order79': {'Product': 'Mouse', 'Price': 25, 'Quantity': 1, 'Delivered': True},
    'order80': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 3, 'Delivered': False},
    'order81': {'Product': 'Monitor', 'Price': 300, 'Quantity': 2, 'Delivered': True},
    'order82': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': False},
    'order83': {'Product': 'Chair', 'Price': 85, 'Quantity': 5, 'Delivered': True},
    'order84': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': False},
    'order85': {'Product': 'Mouse', 'Price': 25, 'Quantity': 4, 'Delivered': True},
    'order86': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 2, 'Delivered': True},
    'order87': {'Product': 'Monitor', 'Price': 300, 'Quantity': 3, 'Delivered': False},
    'order88': {'Product': 'Desk', 'Price': 200, 'Quantity': 2, 'Delivered': True},
    'order89': {'Product': 'Chair', 'Price': 85, 'Quantity': 1, 'Delivered': False},
    'order90': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 4, 'Delivered': True},
    'order91': {'Product': 'Mouse', 'Price': 25, 'Quantity': 2, 'Delivered': True},
    'order92': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 5, 'Delivered': False},
    'order93': {'Product': 'Monitor', 'Price': 300, 'Quantity': 1, 'Delivered': True},
    'order94': {'Product': 'Desk', 'Price': 200, 'Quantity': 3, 'Delivered': False},
    'order95': {'Product': 'Chair', 'Price': 85, 'Quantity': 2, 'Delivered': True},
    'order96': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 3, 'Delivered': False},
    'order97': {'Product': 'Mouse', 'Price': 25, 'Quantity': 1, 'Delivered': True},
    'order98': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 4, 'Delivered': True},
    'order99': {'Product': 'Monitor', 'Price': 300, 'Quantity': 2, 'Delivered': False},
    'order100': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': True},
    'order101': {'Product': 'Chair', 'Price': 85, 'Quantity': 3, 'Delivered': False},
    'order102': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 1, 'Delivered': True},
    'order103': {'Product': 'Mouse', 'Price': 25, 'Quantity': 5, 'Delivered': False},
    'order104': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 2, 'Delivered': True},
    'order105': {'Product': 'Monitor', 'Price': 300, 'Quantity': 4, 'Delivered': True},
    'order106': {'Product': 'Desk', 'Price': 200, 'Quantity': 1, 'Delivered': False},
    'order107': {'Product': 'Chair', 'Price': 85, 'Quantity': 2, 'Delivered': True},
    'order108': {'Product': 'Laptop', 'Price': 1500, 'Quantity': 3, 'Delivered': False},
    'order109': {'Product': 'Mouse', 'Price': 25, 'Quantity': 1, 'Delivered': True},
    'order110': {'Product': 'Keyboard', 'Price': 45, 'Quantity': 5, 'Delivered': False},
}

from collections import defaultdict
from concurrent.futures import ThreadPoolExecutor
from typing import Dict, Tuple, Optional, Union
import logging

def analyze_sales(data: Dict[str, Dict[str, Union[str, float, int, bool]]]) -> Tuple[float, Tuple[Optional[str], float], int]:
    """
    Analyze sales data to compute total revenue, best product by revenue, and pending orders.

    Args:
        data: Dictionary containing sales data with order details

    Returns:
        Tuple containing:
        - Total revenue from delivered orders
        - Tuple of (best product name, its revenue)
        - Number of pending orders

    Raises:
        ValueError: If data format is invalid
    """
    # Input validation
    if not isinstance(data, dict):
        logging.error("Input must be a dictionary")
        raise ValueError("Input must be a dictionary")

    # Initialize defaultdict for automatic handling of missing keys
    product_revenue = defaultdict(float)
    total_revenue = 0.0
    pending_orders = 0

    def process_order(order: Tuple[str, Dict[str, Union[str, float, int, bool]]]) -> Tuple[float, Dict[str, float], int]:
        """Process a single order and return its contribution to metrics."""
        order_id, details = order
        try:
            revenue = float(details['Price']) * int(details['Quantity'])
            is_delivered = bool(details['Delivered'])
            return (
                revenue if is_delivered else 0.0,
                {details['Product']: revenue} if is_delivered else {},
                0 if is_delivered else 1
            )
        except (KeyError, TypeError, ValueError) as e:
            logging.warning(f"Invalid order data for {order_id}: {str(e)}")
            return 0.0, {}, 0

    # Use ThreadPoolExecutor for parallel processing of large datasets
    with ThreadPoolExecutor() as executor:
        results = executor.map(process_order, data.items())

        # Aggregate results
        for rev, prod_rev, pend in results:
            total_revenue += rev
            for prod, rev in prod_rev.items():
                product_revenue[prod] += rev
            pending_orders += pend

    # Find best product efficiently
    best_product = (None, 0.0)
    if product_revenue:
        best_product = max(product_revenue.items(), key=lambda x: x[1])

    return total_revenue, best_product, pending_orders

# Example usage
if __name__ == "__main__":
    logging.basicConfig(level=logging.INFO)
    try:
        total, (product, revenue), pending = analyze_sales(sales_data)
        print(f"Total revenue from delivered orders: ${total:,.2f}")
        print(f"Best product by revenue: {product or 'None'} (${revenue:,.2f})")
        print(f"Number of pending orders: {pending}")
    except ValueError as e:
        print(f"Error: {e}")
```

# 2D Mesh generation üå±

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from tabulate import tabulate

def generate_quad_mesh(x, d, max_elements=1000):
    """
    Generate a quad mesh for 2D soil domain from -x to x in width and 0 to d in depth.

    Parameters:
    - x: half-width (domain extends from -x to x)
    - d: depth (domain from 0 to d)
    - max_elements: approx max number of elements to control mesh density

    Returns:
    - nodes: dict {node_id: (x, y)}
    - elements: dict {element_id: {'nodes': [n1,n2,n3,n4], 'soil_tag': soil_name}}
    - soil_properties: dict of soil layers and properties
    - soil_layer_tags: dict {soil_name: numeric_tag}
    """

    width = 2 * x
    ratio = width / d

    # Determine mesh resolution (n_x, n_y) to approx max_elements total
    n_y = int(np.sqrt(max_elements / ratio))
    n_x = int(ratio * n_y)
    n_x = max(1, n_x)
    n_y = max(1, n_y)

    # Node coordinates
    x_coords = np.linspace(-x, x, n_x + 1)
    y_coords = np.linspace(0, d, n_y + 1)

    nodes = {}
    node_id = 1
    for j, y_val in enumerate(y_coords):
        for i, x_val in enumerate(x_coords):
            nodes[node_id] = (x_val, y_val)
            node_id += 1

    # Define soil layers and properties
    soil_properties = {
        'SoilLayer1': {
            'depth': (0.0, d * 0.4),
            'density': 1800,
            'G': 30e6,
            'nu': 0.3,
            'C': 0,
            'Phi': 30,
            'permeability': 1e-6
        },
        'SoilLayer2': {
            'depth': (d * 0.2, d * 0.6),
            'density': 2000,
            'G': 50e6,
            'nu': 0.25,
            'C': 0.43e2,
            'Phi': 35,
            'permeability': 5e-7
        },
        'SoilLayer3': {
            'depth': (d * 0.6, d),
            'density': 2200,
            'G': 85e6,
            'nu': 0.33,
            'C': 1.5e2,
            'Phi': 27,
            'permeability': 2e-8
        }
    }

    # Assign numeric tags for soil layers
    soil_layer_tags = {name: idx + 1 for idx, name in enumerate(soil_properties)}

    elements = {}
    element_id = 1
    for j in range(n_y):
        for i in range(n_x):
            n1 = j * (n_x + 1) + i + 1
            n2 = n1 + 1
            n3 = n2 + (n_x + 1)
            n4 = n1 + (n_x + 1)

            y_nodes = [nodes[n1][1], nodes[n2][1], nodes[n3][1], nodes[n4][1]]
            elem_depth = sum(y_nodes) / 4

            soil_tag = None
            for soil_name, props in soil_properties.items():
                elev1, elev2 = props['depth']
                if elev1 <= elem_depth < elev2:
                    soil_tag = soil_name
                    break
            if soil_tag is None:
                soil_tag = 'Unknown'

            elements[element_id] = {
                'nodes': [n1, n2, n3, n4],
                'soil_tag': soil_tag
            }
            element_id += 1

    return nodes, elements, soil_properties, soil_layer_tags

def export_to_excel(filename, nodes, elements, soil_properties):
    # Prepare Nodes DataFrame
    nodes_df = pd.DataFrame([
        {'NodeID': nid, 'X': coord[0], 'Y': coord[1]}
        for nid, coord in nodes.items()
    ])

    # Prepare Elements DataFrame
    elements_df = pd.DataFrame([
        {
            'ElementID': eid,
            'N1': data['nodes'][0],
            'N2': data['nodes'][1],
            'N3': data['nodes'][2],
            'N4': data['nodes'][3],
            'SoilTag': data['soil_tag']
        }
        for eid, data in elements.items()
    ])

    # Prepare Soil Properties DataFrame
    soil_df = pd.DataFrame([
        {
            'SoilName': name,
            'DepthStart': props['depth'][0],
            'DepthEnd': props['depth'][1],
            'Density': props['density'],
            'G': props['G'],
            'Nu': props['nu'],
            'C': props['C'],
            'Phi': props['Phi'],
            'Permeability': props['permeability']
        }
        for name, props in soil_properties.items()
    ])

    # Write to Excel with multiple sheets
    with pd.ExcelWriter(filename) as writer:
        nodes_df.to_excel(writer, sheet_name='Nodes', index=False)
        elements_df.to_excel(writer, sheet_name='Elements', index=False)
        soil_df.to_excel(writer, sheet_name='SoilProperties', index=False)

    print(f"Data exported successfully to '{filename}'")


def plot_mesh(nodes, elements, soil_layer_tags, soil_layer_colors):
    fig, ax = plt.subplots(figsize=(10, 5))
    patches = []
    facecolors = []

    for eid, data in elements.items():
        node_ids = data['nodes']
        coords = [nodes[nid] for nid in node_ids]

        polygon = Polygon(coords, closed=True)
        patches.append(polygon)

        soil_name = data['soil_tag']
        color = soil_layer_colors.get(soil_name, "#eeeeee")  # light gray fallback
        facecolors.append(color)

    collection = PatchCollection(patches, facecolors=facecolors, edgecolors='k', linewidths=0.5, alpha=0.5)
    ax.add_collection(collection)

    # Set axis limits explicitly based on node coordinates
    x_vals = [coord[0] for coord in nodes.values()]
    y_vals = [coord[1] for coord in nodes.values()]
    margin_x = (max(x_vals) - min(x_vals)) * 0.05
    margin_y = (max(y_vals) - min(y_vals)) * 0.05

    ax.set_xlim(min(x_vals) - margin_x, max(x_vals) + margin_x)
    ax.set_ylim(min(y_vals) - margin_y, max(y_vals) + margin_y)

    # Create legend handles
    from matplotlib.lines import Line2D
    legend_handles = []
    for soil_name, tag in soil_layer_tags.items():
        legend_handles.append(Line2D([0], [0], marker='s', color='w',
                                     markerfacecolor=soil_layer_colors[soil_name],
                                     markersize=12, label=f"{tag}: {soil_name}"))

    ax.legend(handles=legend_handles, title="Soil Layers", loc='upper right')

    ax.set_aspect('equal')
    ax.set_xlabel("X (m)")
    ax.set_ylabel("Depth (m)")
    ax.set_title("2D Quad Mesh Colored by Soil Layers")
    ax.invert_yaxis()
    ax.grid(True, linestyle='--', linewidth=0.3)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    x = 50  # half width
    d = 45   # depth
    max_elements = 2000

    nodes, elements, soil_props, soil_tags = generate_quad_mesh(x, d, max_elements)

    export_to_excel("mesh_data.xlsx", nodes, elements, soil_props)
    '''
    # Print info in tables
    print(f"Number of nodes: {len(nodes)}")
    print(f"Number of elements: {len(elements)}\n")

    node_sample = [(nid, *coord) for nid, coord in list(nodes.items())[:-5]]
    print("Sample Nodes (first 5):")
    print(tabulate(node_sample, headers=["Node ID", "X", "Y"], tablefmt="grid"))

    elem_sample = [(eid, *edata['nodes'], edata['soil_tag']) for eid, edata in list(elements.items())[:-3]]
    print("\nSample Elements (first 3):")
    print(tabulate(elem_sample, headers=["Element ID", "N1", "N2", "N3", "N4", "Soil Tag"], tablefmt="grid"))

    soil_table = []
    for name, props in soil_props.items():
        soil_table.append([
            name,
            f"{props['depth'][0]:.2f} to {props['depth'][1]:.2f}",
            props['density'],
            props['G'],
            props['nu'],
            props['C'],
            props['Phi'],
            props['permeability']
        ])
    print("\nSoil Properties:")
    print(tabulate(soil_table, headers=["Soil Name", "Depth Range", "Density", "G", "nu", "C", "Phi", "Permeability"], tablefmt="grid"))
    '''

    # Assign soft colors for soil layers
    soft_colors = [
        "#e0d0f9",  # light purple
        "#d0f9d9",  # light green
        "#f9f1d0",  # light yellow
        "#d0e1f9",  # light blue
        "#f9d0d0",  # light pink
    ]
    soil_colors = {name: soft_colors[i % len(soft_colors)] for i, name in enumerate(soil_props)}

    # Plot mesh with colored soil layers
    plot_mesh(nodes, elements, soil_tags, soil_colors)
```
