```python
def is_even(n):
    return n % 2 == 0

def pick_evens(*args):
    nums = []
    for i in args:
        if is_even(i):
            nums.append(i)
    return nums

pick_evens(7, 13, 19, 21)
```

# Try and Exceptions ğŸ¦‰

```python
def divis(a, b):
    r = None
    try:
        r = a / b
    except ZeroDivisionError:
        print('could not divide by zero')
        r = None
    except Exception as e:
        print(f'had error {e}')
        r = None
    finally:
        return r

divis(1, 0)
```

# Decorators ğŸ

```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
class C:
    def __init__(self):
        self._x = None

    def getx(self):
        return self._x

    def setx(self, value):
        self._x = value

    def delx(self):
        del self._x

    x = property(getx, setx, delx, "I'm the 'x' property.")
```

```python
def make_sound(func):
    def wrapper():
        user_sound = func()
        return user_sound + "this is a sample!"
    return wrapper

@make_sound
def sound():
    return str(1 / 5) + " "

print(sound())
```

```python
# importing the module
import functools

# decorator
def make_geek_happy(func):
    @functools.wraps(func)
    def wrapper():
        neutral_message = func()
        happy_message = neutral_message + " You are happy!"
        return happy_message
    return wrapper

def speak():
    """Returns a neutral message"""
    return "Hi!"

positive_message = make_geek_happy(speak)
#print(positive_message())

print("name is: ", speak.__name__)
print("comment is: ", speak.__doc__)
```

```python
def calc_time(func):
    def wrapper(*args, **kwargs):
        import time
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        elapsed_time = end_time - start_time
        print(f"Elapsed time: {elapsed_time:.6f} seconds")
        return result
    return wrapper

@calc_time
def make_list(n):
    list = [x for x in range(n)]
    return list

print(make_list(int(input("give me a number: "))))
```

```python
import csv

with open("products.csv", "r", encoding="utf-8") as infile:
    reader = csv.DictReader(infile)
    rows = []

    for row in reader:
        name = row["Product Name"]
        price = int(row["Price"])
        qty = int(row["Quantity"])
        total_price = price * qty

        # Ø§Ø¶Ø§ÙÙ‡ Ú©Ø±Ø¯Ù† Ø³ØªÙˆÙ† Ø¬Ø¯ÛŒØ¯
        row["Total Price"] = total_price
        rows.append(row)

with open("products_output.csv", "w", encoding="utf-8", newline="") as outfile:
    fieldnames = ["Product Name", "Price", "Quantity", "Total Price"]
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(rows)

print("âœ… ÙØ§ÛŒÙ„ products_output.csv Ø³Ø§Ø®ØªÙ‡ Ø´Ø¯")
```
