```python
import requests
from bs4 import BeautifulSoup
```

Getting the html code of blah blah blah site.

```python
r = requests.get("https://waitbutwhy.com/2016/09/marriage-decision.html")
print(f"response code is:", r.status_code)
data = r.content
```

parsing the response

```python
page = BeautifulSoup(data, "html.parser")
text_data = page.select_one('#post-5618 > div')

# Print the result if found
if text_data:
    print(text_data.get_text(strip=False))
else:
    print("Element not found.")
```
